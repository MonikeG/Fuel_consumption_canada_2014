#### Seleção dos dados para exploração

```python
consumo_comb = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB_MPG']]
consumo_comb.head()
```

#### Dataset Treino e teste

```python
#create a mask to select random rows using np.random.rand() function:
mask = np.random.rand(len(consumo_comb))<0.8
train = consumo_comb[mask]
test = consumo_comb[mask]
```

### Regressão linear simples - Treino

```python
from sklearn import linear_model
regressao = linear_model.LinearRegression()
train_x = np.asanyarray(train[["ENGINESIZE"]])
train_y = np.asanyarray(train[["CO2EMISSIONS"]])
regressao.fit(train_x, train_y)
print("Coeficiente angular:", regressao.coef_)
print("Coeficiente linear:", regressao.intercept_)

```

```python
plt.scatter(consumo_comb.ENGINESIZE, consumo_comb.CO2EMISSIONS, color = '#D46A6A')
plt.plot(train_x, regressao.coef_[0][0]*train_x + regressao.intercept_[0], '-r')
plt.xlabel("ENGINESIZE")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

### Avaliação do modelo

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(train[["ENGINESIZE"]])
test_y = np.asanyarray(train[["CO2EMISSIONS"]])
test_y_= regressao.predict(test_x)
```

```python
print("Erro médio absoluto : %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Resíduo (MSE): %.2f" % np.mean((test_y_ - test_y)**2))
print("Rquadrado: %.2f" % r2_score(test_y, test_y_))
```

### Regressão Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

train_x = np.asanyarray(train[["ENGINESIZE"]])
train_y = np.asanyarray(train[["CO2EMISSIONS"]])

test_x = np.asanyarray(test[["ENGINESIZE"]])
test_y = np.asanyarray(test[["CO2EMISSIONS"]])

polinomial = PolynomialFeatures(degree=2)
train_x_poli = polinomial.fit_transform(train_x)
train_x_poli
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poli, train_y)

print("Coeficiente angular:", clf.coef_)
print("Coeficiente linear:", clf.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color = '#D46A6A')

XX = np.arange(0.0, 10.0, 0.1)

YY = clf.intercept_[0] + clf.coef_[0][1]*XX + clf.coef_[0][2]*np.power(XX, 2)

plt.plot(XX, YY, '-r')
plt.xlabel("ENGINESIZE")
plt.ylabel("EMISSION")
```

```python
test_x_poli = polinomial.transform(test_x)
test_y = clf.predict(test_x_poli)

print("Erro médio absoluto : %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Resíduo (MSE): %.2f" % np.mean((test_y_ - test_y)**2))
print("Rquadrado: %.2f" % r2_score(test_y, test_y_))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model


polinomial3 = PolynomialFeatures(degree=3)
train_x_poli3 = polinomial3.fit_transform(train_x)

clf3 = linear_model.LinearRegression()
train_y3_ = clf3.fit(train_x_poli3, train_y)

print("Coeficiente angular:", clf3.coef_)
print("Coeficiente linear:", clf3.intercept_)

plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color = '#D46A6A')

XX = np.arange(0.0, 10.0, 0.1)

YY = clf3.intercept_[0] + clf3.coef_[0][1]*XX + clf3.coef_[0][2]*np.power(XX, 2) + clf3.coef_[0][3]*np.power(XX, 3)

plt.plot(XX, YY, '-r')
plt.xlabel("ENGINESIZE")
plt.ylabel("EMISSION")

test_x_poli3 = polinomial3.transform(test_x)
test_y3_ = clf3.predict(test_x_poli3)

print("Erro médio absoluto : %.2f" % np.mean(np.absolute(test_y3_ - test_y)))
print("Resíduo (MSE): %.2f" % np.mean((test_y3_ - test_y)**2))
print("Rquadrado: %.2f" % r2_score(test_y, test_y3_))
```
